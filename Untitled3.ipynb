```python
import numpy as np

# Load daily closing prices
prices = data["Close"].values.flatten()

# Parameters
n_states = 3
n_actions = 3  # 0 = Hold, 1 = Buy, 2 = Sell
Q = np.zeros((n_states, n_actions))

alpha = 0.1
gamma = 0.95
epsilon = 0.30  # Increased initial exploration
episodes = 2000

# Define threshold for flat/sideways state
THRESHOLD = 0.0015

def get_state(pct_change):
    if pct_change > THRESHOLD:
        return 2  # Uptrend
    elif pct_change < -THRESHOLD:
        return 0  # Downtrend
    else:
        return 1  # Sideways / Flat

# Train RL Agent
for episode in range(episodes):
    holding = False
    buy_price = 0
    
    # Decaying epsilon over training
    current_epsilon = max(0.01, epsilon * (1 - episode / episodes))
    
    for i in range(1, len(prices) - 1):
        pct_change = (prices[i] - prices[i - 1]) / prices[i - 1]
        state = get_state(pct_change)
        
        # Epsilon-greedy selection
        if np.random.rand() < current_epsilon:
            action = np.random.randint(n_actions)
        else:
            action = np.argmax(Q[state])
            
        reward = 0
        
        if action == 1:  # BUY
            if not holding:
                holding = True
                buy_price = prices[i]
                reward = 0
            else:
                reward = -0.1  # Moderate penalty for duplicate buy
                
        elif action == 2:  # SELL
            if holding:
                # Scale return by 100 so rewards match the penalty scale
                trade_return = (prices[i] - buy_price) / buy_price
                reward = trade_return * 100
                holding = False
            else:
                reward = -0.1  # Moderate penalty for invalid sell
                
        else:  # HOLD
            if holding:
                # Small step reward for holding during price increases
                step_return = (prices[i] - prices[i - 1]) / prices[i - 1]
                reward = step_return * 100
            else:
                reward = -0.01  # Tiny penalty for remaining uninvested
                
        # Next state evaluation
        next_pct = (prices[i + 1] - prices[i]) / prices[i]
        next_state = get_state(next_pct)
        
        # Bellman Update
        Q[state, action] += alpha * (
            reward + gamma * np.max(Q[next_state]) - Q[state, action]
        )

print("Learned Q-table:")
print(Q)
```

```python
# Test RL Strategy (Fixed)
capital = 10000
shares = 0
holding = False
portfolio = []

for i in range(1, len(prices)):
    # Calculate percentage change (matches training logic)
    pct_change = (prices[i] - prices[i - 1]) / prices[i - 1]
    state = get_state(pct_change)
    
    action = np.argmax(Q[state])
    
    # BUY
    if action == 1 and not holding:
        shares = capital / prices[i]
        capital = 0
        holding = True
    # SELL
    elif action == 2 and holding:
        capital = shares * prices[i]
        shares = 0
        holding = False
        
    value = capital if not holding else shares * prices[i]
    portfolio.append(value)

portfolio = np.array(portfolio)
```

```python
# Performance Metrics
def performance(portfolio):
    # Force the portfolio array to be completely flat (1D)
    portfolio = np.array(portfolio).flatten() 
    
    returns = pd.Series(portfolio).pct_change().dropna()
    total_return = (portfolio[-1] - portfolio[0]) / portfolio[0]
    annual_return = (1 + total_return) ** (252 / len(portfolio)) - 1
    volatility = returns.std() * np.sqrt(252)
    sharpe = annual_return / volatility
    running = np.maximum.accumulate(portfolio)
    drawdown = (portfolio - running) / running
    max_dd = drawdown.min()
    
    return {
        "Total Return": total_return, 
        "Annual Return": annual_return, 
        "Volatility": volatility, 
        "Sharpe": sharpe, 
        "Max Drawdown": max_dd
    }
```

```python
# LSTM Strategy (Fixed Signal Logic)
actual_flat = actual.flatten()
predicted_flat = predicted.flatten()

capital = 10000
shares = 0
holding = False
values = []

for i in range(len(predicted_flat) - 1):
    # Predict trend direction using LSTM's relative change (Tomorrow vs Today)
    pred_trend_up = predicted_flat[i + 1] > predicted_flat[i]
    
    # BUY Signal
    if pred_trend_up and not holding:
        shares = capital / actual_flat[i]
        capital = 0
        holding = True
        
    # SELL Signal
    elif not pred_trend_up and holding:
        capital = shares * actual_flat[i]
        shares = 0
        holding = False
        
    # Mark-to-market daily portfolio value
    current_val = capital if not holding else shares * actual_flat[i]
    values.append(current_val)

# Final day valuation
final_val = capital if not holding else shares * actual_flat[-1]
values.append(final_val)

values = np.array(values)
```
